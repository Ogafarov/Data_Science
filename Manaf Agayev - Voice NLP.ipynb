```python
import librosa 
```

```python
audio_path = r"C:\Users\orkha\OneDrive\Desktop\Python\Python\August\Manaf_Agayev_-_Ay_cani_yanmis_-_www.BiG.AZ.mp3"
```

```python
y,sr = librosa.load(audio_path,sr=None)
```

```python
print(f"Sesin numune derecesi: {sr}")
print(f"sesin uzunlugu:{len(y)}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,5))
plt.plot(y)
plt.title("ses siqnallari")
plt.xlabel("Zaman")
plt.ylabel("Amplitud")
plt.show()
```

```python
import numpy as np 
import librosa.display 
S=librosa.feature.melspectrogram(y=y,sr=sr, n_mels = 128)
S_dB = librosa.power_to_db(S, ref=np.max)
plt.figure(figsize=(14,5))
librosa.display.specshow(S_dB,sr=sr,x_axis="time",y_axis='mel')
plt.colorbar(format = "%+2.0f dB")
plt.title("melspektogram")
plt.xlabel("zaman(saniye)")
plt.ylabel("Mel Frekanslarini")
plt.show()
```
