```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV

wine = load_wine()
X, y = wine.data, wine.target 
df = pd.DataFrame(X, columns=wine.feature_names)
df["target"] = y
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
param_grid = {"n_neighbors": np.arange(1,31), "weights":["uniform","distance"]}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid,cv=5)
grid_search.fit(X_train_scaled, y_train)
print("\nEn yaxsi param: ", grid_search.best_params_)
print("en yaxsi skor: ",grid_search.best_score_)
```

```python
k_range = range(1,20)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled,y_train)
    k_scores.append(knn.score(X_test_scaled,y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(k_range,k_scores)
plt.xlabel("K deyeri")
plt.ylabel("test deqiqliyi")
plt.show()
```
